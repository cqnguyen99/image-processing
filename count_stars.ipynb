```python
import os 
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from math import sqrt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray
from skimage.io import imread
import glob
```

```python
os.chdir('/Users/q.chi1099/Desktop/')
```

```python
example_image = glob.glob('night_sky.gif')[0]
im = imread(example_image, as_gray=False)
plt.imshow(im, cmap=cm.gray)
plt.show()
```

```python
image_gray = rgb2gray(im)  # Change image to gray

blobs_log = blob_log(image_gray, max_sigma=30, num_sigma=10, threshold=.1)

# Compute radii in the 3rd column
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
num_row = len(blobs_log)
print("Numbers of stars counted: ", num_row)

fig, axes = plt.subplots(1,1)
plt.imshow(image_gray, cmap=cm.gray)

for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x,y), r+5, color='salmon', linewidth=1, fill=False)
    axes.add_patch(c)
```
